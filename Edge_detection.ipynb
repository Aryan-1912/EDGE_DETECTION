```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./Giraffe.jpg", cv2.IMREAD_GRAYSCALE)

# sobel edge 

sobel_x= cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 3 )
sobel_y= cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize = 3 )

sobel = cv2.magnitude(sobel_x, sobel_y)


# laplacian
# moreover laplacian gives the points where edges 
# are there regardless of the orientation so we will get 
# 2 points in total


laplacian = cv2.Laplacian(image, cv2.CV_64F,  ksize = 3 )

# lapcian abs only have + values 
laplacian_abs = cv2.convertScaleAbs(laplacian)

# canny


canny_images = cv2.Canny(image, 180, 200)


titles = ['Original', 'Sobel','laplacian_all', 'laplacian_abs', 'canny']

images = [image , sobel, laplacian, laplacian_abs, canny_images ]

plt.figure(figsize=(12,8))

for i in range(len(titles)):
    plt.subplot(3,2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')


plt.tight_layout()
plt.show()





```
